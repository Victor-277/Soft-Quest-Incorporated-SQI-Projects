```python
import numpy as np 
import pandas as pd 
import matplotlib 
import matplotlib.pyplot as plt 
import sklearn
print(sklearn.__version__)
print(np.__version__)
```

```python
import seaborn as sns
# tips = sns.load_dataset('tips')
# tips.head()
tips = pd.read_csv("C:\\Machine learning\\tips.csv")
tips.head()
```

```python
print(tips.shape)
tips.gender.unique()
```

```python
tips["gender"] = tips.gender.replace({"Female":0,"Male":1})
tips.head()
```

```python
tips.smoker.unique()
```

```python
tips["smoker"] = tips.smoker.replace({"No":0,"Yes":1})
tips.head()
```

```python
tips.day.unique()
```

```python
tips["day"] = tips.day.replace({"Sun":1,"Sat":7,"Thur":5,"Fri":6})
tips.head()
```

```python
tips.time.unique()
```

```python
tips["time"] = tips.time.replace({"Dinner":0,"Lunch":1})
tips.head()
```

```python
%matplotlib inline
sns.set()
sns.pairplot(tips, height=1.5);
```

```python
tips.drop(['gender','day','smoker','time'], axis=1,inplace=True)
tips.head()
```

```python
X_tips = tips.drop('total_bill', axis=1)
X_tips.shape
```

```python
y_tips = tips['total_bill']
y_tips.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_tips, y_tips, random_state=1, train_size=0.8, test_size=0.2)
```

```python
Xtrain.shape
```

```python
# 1. Choose a class of model.
from sklearn.linear_model import LinearRegression
# 2. Choose model hyperparameters
model = LinearRegression(fit_intercept=True)
model
# LinearRegression?
```

```python
model.fit(Xtrain, ytrain) # 3. fit model to data
y_model = model.predict(Xtest)
```

```python
print("Predicted \n", y_model)
print("Original label \n", ytest.values)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot
plt.scatter(ytest, y_model, color='blue', alpha=0.6, label="Predicted vs Actual")

# Perfect Prediction Line (y = x)
x_line = np.linspace(min(ytest), max(ytest), 100)
plt.plot(x_line, x_line, color='red', linestyle='dashed', label="Perfect Fit (y = x)")

# Labels and title
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.legend()
plt.show()
```

```python
import seaborn as sns

residuals = ytest - y_model

# Plot residuals
sns.histplot(residuals, bins=30, kde=True)
plt.xlabel("Residuals (Actual - Predicted)")
plt.title("Residuals Distribution")
plt.show()
```
