```python
 from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup as soup
import pandas as pd
import numpy as np
import time

# Specify the number of pages to scrape
pages_to_scrape = 100

# Initialize Selenium WebDriver with custom User-Agent
from selenium.webdriver.chrome.options import Options
options = Options()
options.add_argument("--headless")
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
)
browser = webdriver.Chrome(executable_path=r"C:\chromedriver.exe", options=options)
browser.get("https://www.jumia.com.ng/catalog/?q=laptop")

# Initialize empty lists for product details
products_image = []
description = []
products_rating = []
new_pricing = []
old_pricing = []
discount = []

# Loop through the specified number of pages
current_page = 1
while current_page <= pages_to_scrape:
    print(f"Scraping page {current_page} of {pages_to_scrape}...")
    try:
        # Wait for the product containers to load
        WebDriverWait(browser, 10).until(
            EC.presence_of_element_located((By.XPATH, "//article[@class='prd _fb col c-prd']"))
        )
        
        # Get the page source
        html_source = browser.page_source
        page_soup = soup(html_source, "html.parser")
        
        # Find the product container
        product_container = page_soup.findAll("article", {"class": "prd _fb col c-prd"})
        
        for product in product_container:
            # Scrape product image
            image = product.findAll("div", {"class": "img-c"})
            if len(image) == 0 or not image[0].find("img"):
                products_image.append(np.nan)
            else:
                img_tag = image[0].find("img")
                if "data-src" in img_tag.attrs:
                    products_image.append(img_tag["data-src"])
                elif "src" in img_tag.attrs:
                    products_image.append(img_tag["src"])
                else:
                    products_image.append(np.nan)
            
            # Scrape product description
            describe = product.findAll("h3", {"class": "name"})
            if len(describe) == 0:
                description.append(np.nan)
            else:
                description.append(describe[0].text.strip())
            
            # Scrape rating and review
            rating = product.findAll("div", {"class": "stars _s"})
            if len(rating) == 0:
                products_rating.append(np.nan)
            else:
                products_rating.append(rating[0].text.strip())

            # Scrape new price
            new_price = product.findAll("div", {"class": "prc"})
            if len(new_price) == 0:
                new_pricing.append(np.nan)
            else:
                new_pricing.append(new_price[0].text.strip())
            
            # Scrape old price
            old_price = product.findAll("div", {"class": "old"})
            if len(old_price) == 0:
                old_pricing.append(np.nan)
            else:
                old_pricing.append(old_price[0].text.strip())

            # Scrape discount
            price_cut = product.findAll("div", {"class": "bdg _dsct _sm"})
            if len(price_cut) == 0:
                discount.append(np.nan)
            else:
                discount.append(price_cut[0].text.strip())
        
        # Check if the "Next" button exists and go to the next page
        try:
            # Wait for the "Next Page" button to be clickable
            next_button = WebDriverWait(browser, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[@aria-label='Next Page']"))
            )
            
            # Scroll to the bottom of the page to make sure the button is visible
            browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1)  # Wait for the scroll to finish
            
            # Use JavaScript to click the button
            browser.execute_script("arguments[0].click();", next_button)
            time.sleep(2)  # Wait for the next page to load
            current_page += 1

        except NoSuchElementException:
            print("No more pages to scrape.")
            break
        except Exception as e:
            print(f"Error while clicking next page: {e}")
            break

    except TimeoutException:
        print(f"Timeout while loading page {current_page}.")
        break
    except Exception as e:
        print(f"Error on page {current_page}: {e}")
        break

# Close the browser
browser.quit()

# Create a DataFrame with the scraped data
df = pd.DataFrame({
    "Description": description,
    "Rating": products_rating,
    "Image": products_image,
    "Discount": discount,
    "Original Price": old_pricing,
    "Discounted Price": new_pricing,
})

# Remove duplicates
df = df.drop_duplicates()

# Save the DataFrame to a CSV file
df.to_csv("jumia_laptops.csv", index=False)

# Display the first few rows of the scraped data
print(df.head(30))
```
