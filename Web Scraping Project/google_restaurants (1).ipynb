```python
import pandas as pd
import numpy as np
from selenium import webdriver  
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys  
from selenium.common.exceptions import NoSuchElementException  
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup as soup

# Initialize the WebDriver correctly
service = Service(ChromeDriverManager().install())  # Automatically downloads/upgrades ChromeDriver
browser = webdriver.Chrome(service=service)

# Open the Google Maps page for restaurants in New York City
browser.get("https://www.google.com/maps/search/restaurants+in+New+York+City,+NY,+USA/@40.760001,-74.0557728,13z?entry=ttu&g_ep=EgoyMDI1MDQwNi4wIKXMDSoASAFQAw%3D%3D") 

# List to store restaurant URLs
restaurant_url = []

# Scroll and collect restaurant URLs
scroll_count = 0
max_scrolls = 10  # Limit to 10 links for testing

while len(restaurant_url) < max_scrolls:
    # Wait for the page to load more results after each scroll
    time.sleep(7)
    
    # Get the page source after scrolling
    html_source = browser.page_source  
    page_soup = soup(html_source, 'html.parser')
    
    # Find containers that hold restaurant links (you may adjust this based on the class name)
    containers = page_soup.findAll('div', {'class': 'Nv2PK THOPZb CpccDe'})
    
    # Loop through containers and extract the URLs
    for container in containers:
        try:
            url = container.find("a", {"class": "hfpxzc"})["href"]
            if url not in restaurant_url:  # Avoid duplicate URLs
                restaurant_url.append(url)
        except (IndexError, TypeError):
            continue  # Skip if there's no link found
    
    # Scroll down the page to load more restaurants
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    scroll_count += 1
    if scroll_count >= max_scrolls:
        break

# Print the collected restaurant URLs (for testing)
for url in restaurant_url:
    print(url)

# Save URLs to a CSV file for later use (optional)
df = pd.DataFrame(restaurant_url, columns=["Restaurant_URL"])
df.to_csv("restaurant_urls.csv", index=False)

# Close the browser after scraping
browser.quit()
```

```python
import pandas as pd
import re
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from bs4 import BeautifulSoup as soup

# Set up Selenium WebDriver
service = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=service)

#Get Restaurant URLs
search_url = "https://www.google.com/maps/search/restaurants+in+New+York+City,+NY,+USA/"
browser.get(search_url)
time.sleep(5)

restaurant_url = []
scroll_count = 0
max_scrolls = 100  # Adjust to your need

while scroll_count < max_scrolls:
    # Scroll the listings container
    try:
        scrollable_div = browser.find_element(By.XPATH, '//div[@role="feed"]')
        browser.execute_script("arguments[0].scrollBy(0, 1000);", scrollable_div)
    except Exception as e:
        print("Scrolling failed:", e)
        break

    time.sleep(5)

    # Find all restaurant place URLs
    try:
        elements = browser.find_elements(By.XPATH, '//a[contains(@href, "/place/")]')
        for elem in elements:
            url = elem.get_attribute("href")
            if url and url not in restaurant_url:
                restaurant_url.append(url)
    except Exception as e:
        print("Error finding links:", e)

    print(f"Collected {len(restaurant_url)} URLs after {scroll_count + 1} scroll(s).")
    scroll_count += 1

# Save the collected URLs
pd.DataFrame(restaurant_url, columns=["Restaurant_URL"]).to_csv("restaurant_urls.csv", index=False)
print(f"Total unique restaurant URLs collected: {len(restaurant_url)}")

browser.quit()  

#Scrape details from each restaurant URL
browser = webdriver.Chrome(service=service)

restaurant_names, addresses, open_hours = [], [], []
websites, phones, ratings = [], [], []
dates, reviews, geo_locations = [], [], []

for index, url in enumerate(restaurant_url):
    print(f"Scraping {index + 1}/{len(restaurant_url)}: {url}")
    browser.get(url)
    time.sleep(5)
    html_source = browser.page_source
    page_soup = soup(html_source, 'html.parser')

    # Name
    try:
        name = page_soup.find("h1", {"class": "DUwDvf lfPIob"}).text.strip()
    except:
        name = "No name found"
    restaurant_names.append(name)

    # Address
    try:
        address = page_soup.find("button", {"class": "CsEnBe"}).text.strip()
    except:
        address = "No address found"
    addresses.append(address)

    # Opening hours
    try:
        week_hours = page_soup.find_all("tr", {"class": "y0skZc"})
        all_days = [day.text.strip() for day in week_hours]
        open_hours.append(" | ".join(all_days))
    except:
        open_hours.append("No opening hours found")

    # Website
    try:
        website_link = page_soup.find_all("a", {"class": "CsEnBe"})
        website = next((a["href"] for a in website_link if "http" in a["href"]), "No website found")
    except:
        website = "No website found"
    websites.append(website)

    # Phone
    try:
        phone = page_soup.find("button", {"data-tooltip": "Copy phone number"}).text.strip()
    except:
        phone = "No phone number found"
    phones.append(phone)

    # Ratings
    try:
        rat = page_soup.find("div", {"class": "fontDisplayLarge"}).text.strip()
    except:
        rat = "No rating"
    ratings.append(rat)

    # Latest Review Date
    try:
        dat = page_soup.find("span", {"class": "rsqaWe"}).text.strip()
    except:
        dat = "No date"
    dates.append(dat)

    # Review
    try:
        rev = page_soup.find("span", {"class": "wiI7pd"}).text.strip()
    except:
        rev = "No review"
    reviews.append(rev)

    # Geo-location (from URL or meta tag)
    try:
        meta = browser.find_element(By.XPATH, '//meta[@itemprop="image"]')
        content = meta.get_attribute("content")
        latlon_match = re.search(r'w\d+-h\d+-k-no-p(?:-v\d+)?-([-\d.]+)_([-\d.]+)', content)

        if latlon_match:
            lat = latlon_match.group(1)
            lon = latlon_match.group(2)
            geo_locations.append(f"{lat}, {lon}")
        else:
            # Try from current URL
            match = re.search(r'@(-?\d+\.\d+),(-?\d+\.\d+)', browser.current_url)
            if match:
                lat, lon = match.groups()
                geo_locations.append(f"{lat}, {lon}")
            else:
                geo_locations.append("No geo-location found")
    except:
        geo_locations.append("No geo-location found")

    print(f"✔️ Done: {name}")

# Save final data to CSV
data = {
    "Name": restaurant_names,
    "Address": addresses,
    "Opening Hours": open_hours,
    "Website": websites,
    "Phone": phones,
    "Ratings": ratings,
    "Review Date": dates,
    "Review": reviews,
    "Geo-Location": geo_locations
}

df = pd.DataFrame(data)
df.to_csv("scraped_restaurants.csv", index=False)
browser.quit()

print("Scraping complete. Data saved to 'scraped_restaurants.csv'")
```
