```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\HR-Employee-Attrition.csv")
df.head(20)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Attrition_Flag'] = df['Attrition'].replace({'Yes':1, 'No': 0})
df.head()
```

### Create new metrics

```python
total_employees = len(df)
attrition_count = df['Attrition_Flag'].sum()
print(f"Total Employees : {total_employees}")
print(f"Attrition Count: {attrition_count}")
```

```python
attrition_rate = round(attrition_count/total_employees * 100,2)
print(f"Attrition Rate : {attrition_rate}")
```

```python
average_tenure = df["YearsAtCompany"].mean().round(2)
average_satisfaction = df["JobSatisfaction"].mean().round(2)
print(f"Average Tenure(Year) : {average_tenure}")
print(f"Average Satisfaction(1-4): {average_satisfaction}")
```

#### 1. Bar Chart – Attrition Count

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Attrition", data=df, palette="Set2")
plt.title("Attrition Count (Yes vs No)")
plt.show()
```

#### 2. Bar Chart – Attrition by Department

```python
plt.figure(figsize=(7,5))
sns.countplot(x="Department", hue="Attrition", data=df, palette="Set1")
plt.title("Attrition by Department")
plt.xticks(rotation=0)
plt.show()
```

#### 3. Bar Chart – Attrition by Job Role

```python
plt.figure(figsize=(12,6))
sns.countplot(
    x="JobRole", 
    hue="Attrition", 
    data=df, 
    palette="Set2",
    order=df["JobRole"].value_counts().index  # sort by frequency
)
plt.title("Attrition by Job Role", fontsize=14, weight='bold')
plt.xlabel("Job Role", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 3. Boxplot – Monthly Income vs Attrition

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df, palette="Pastel1")
plt.title("Monthly Income vs Attrition")
plt.show()
```

### 4. Heatmap – Correlation of Key Features

```python
corr_cols = ["Age", "MonthlyIncome", "YearsAtCompany", "PerformanceRating", "Attrition_Flag"]
plt.figure(figsize=(7,5))
sns.heatmap(df[corr_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### 5. Distribution Plots – WorkLifeBalance & JobSatisfaction

```python
plt.figure(figsize=(6,4))
sns.countplot(x="WorkLifeBalance", data=df, palette="Blues")
plt.title("Distribution of WorkLifeBalance (1-4)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="JobSatisfaction", data=df, palette="Greens")
plt.title("Distribution of Job Satisfaction (1-4)")
plt.show()
```
