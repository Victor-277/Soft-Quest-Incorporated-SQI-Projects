```python
#TO make your plots display within the notebook instead of opening a new window to display the chart
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv(r"C:\ELEVVO\train.csv")
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic.info()
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
titanic.head(5)
```

```python
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic.head(5)
```

```python
titanic.drop(columns='Cabin', inplace=True)
titanic.head(5)
```

```python
titanic.drop(columns=["Ticket", "PassengerId"], inplace=True)
titanic.head(5)
```

```python
titanic['Survived'].replace({0: 'No', 1: 'Yes'}, inplace=True)
titanic.head(5)
```

```python
titanic['Pclass'].unique()
```

```python
titanic['Pclass'].replace({1: 'First Class', 2: 'Second Class', 3: 'Third Class'}, inplace=True)
titanic.head()
```

```python
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
titanic.head()
```

```python
titanic['FamilySize'].unique()
```

```python
def categorize_family(size):
    if size == 1:
        return 'Alone'
    elif size <= 4:
        return 'Small'
    else:
        return 'Large'

titanic['FamilyGroup'] = titanic['FamilySize'].apply(categorize_family)

titanic.head(5)
```

```python
titanic['Embarked'].unique()
```

```python
titanic['Embarked'].replace({"S":"Southampton", "C":"Cherbourg ","Q":"Queenstown"}, inplace=True)
titanic.head(5)
```

```python
titanic.drop(columns=["Parch", "SibSp", "Name"], inplace=True)
titanic.head(5)
```

```python
titanic['FareBand'] = pd.qcut(titanic['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
titanic.head(5)
```

```python
titanic.drop(columns=["Fare"], inplace = True)
titanic.head(5)
```

```python
titanic.to_csv("titanic_eda_data.csv", index = False)
print("Done")
```
