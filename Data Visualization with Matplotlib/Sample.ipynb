```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
df = pd.read_excel(r"C:\Users\DELL\Downloads\Materials_Labour_Q1_2024_2025.xlsx")

# View data
print(df)
```

```python
print(df.describe())
```

```python
import matplotlib.pyplot as plt

materials = df['Material/Trade']
price_2024 = df['Unit Price (Q1 2024, ₦)']/1000   # divide by 1000
price_2025 = df['Unit Price (Q1 2025, ₦)']/1000   # divide by 1000

x = range(len(materials))

plt.bar(x, price_2024, width=0.4, label='Q1 2024', align='center')
plt.bar(x, price_2025, width=0.4, label='Q1 2025', align='edge')

plt.xticks(x, materials, rotation=45)
plt.ylabel("Price (₦)")
plt.title("Materials & Labour Prices Q1 2024 vs Q1 2025")
plt.legend()
plt.tight_layout()
plt.show()
```
