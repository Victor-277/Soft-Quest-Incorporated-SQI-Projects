```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv(r'crop.csv')
df.info()
```

```python
df.describe()
```

```python
df
```

```python
df_cleaned=df.dropna(subset=['SiteID', 'Location'])
df_cleaned
```

```python
df_cleaned
```

```python
df_cleaned['Instrument Type'] = df_cleaned['Instrument Type'].str.replace(r"[()]", "", regex=True)
df_cleaned
```

```python
for column in df_cleaned.columns:
    if df_cleaned[column].dtype in ['float64', 'int64']:
        df_cleaned[column] = df_cleaned[column].fillna(df_cleaned[column].mean())
    elif df_cleaned[column].dtype == 'object':
        df_cleaned[column] = df_cleaned[column].fillna(df_cleaned[column].mode()[0])
    elif df_cleaned[column].dtype == 'bool':
        df_cleaned[column] = df_cleaned[column].fillna(False)
```

```python

df_cleaned['Air Pressure'] = df_cleaned['Air Pressure'].fillna('DefaultValue')
```

```python
df_cleaned
```

```python
df_cleaned.to_csv('clean.csv', sep=',', index=False)
```
